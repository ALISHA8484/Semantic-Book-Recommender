```python
import kagglehub
from seaborn import heatmap

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)
plt.xlabel("Columns")
plt.ylabel("Missed data")
```

```python
books["Missing_description"] = np.where(books["description"].isna(),1 , 0)
books ["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book","Missing_description", "average_rating"]
correlation_Matrix = books[columns_of_interest].corr( method = "spearman")


sns.set_theme(style="white")
plt.figure(figsize = (8,8))
heatmap = sns.heatmap(correlation_Matrix, annot=True, fmt= ".2f",cmap="coolwarm" , cbar_kws={"label": "Spearman Correlation"})

heatmap.set_xlabel("Correlation heatmap")
plt.show()
```

```python
book_missing = books[
    ~(books["description"].isnull()) &
    ~(books["num_pages"].isnull()) &
    ~(books["average_rating"].isnull()) &
    ~(books["published_year"].isnull())
]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
category_counts = book_missing["categories"].value_counts().reset_index()
category_counts.columns = ["category", "count"]
top_n = 20
top_categories = category_counts.head(top_n)

sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

sns.barplot(
    y=top_categories["category"],
    x=top_categories["count"],
    palette="magma"
)

plt.title(f"Top {top_n} Book Categories", fontsize=20)
plt.xlabel("Number of Books", fontsize=15)
plt.ylabel("Category", fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
book_missing["words_in_description"].value_counts().sort_index()


word_counts = book_missing["words_in_description"].value_counts().sort_index()
filtered_counts = word_counts[word_counts.index <= 300]

sns.set(style="whitegrid")
plt.figure(figsize=(18, 7))

sns.barplot(x=filtered_counts.index, y=filtered_counts.values, palette="viridis")

plt.title("Distribution of Description Lengths (Up to 300 Words)", fontsize=20)
plt.xlabel("Number of Words in Description", fontsize=16)
plt.ylabel("Number of Books", fontsize=16)

xticks_to_show = [x for x in filtered_counts.index if x % 20 == 0]
plt.xticks(ticks=xticks_to_show, labels=xticks_to_show, rotation=45, fontsize=12)

plt.yticks(fontsize=13)

plt.tight_layout()
plt.show()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
        np.where(book_missing_25_words["subtitle"].isna(),
                 book_missing_25_words["title"],
                 book_missing_25_words[["title" , "subtitle"]].astype(str).agg(": ".join,axis = 1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13" , "description"]].astype(str).agg(": ".join,axis = 1)
```

```python
(
    book_missing_25_words
    .drop(columns=["Missing_description", "average_rating" , "subtitle" , "words_in_description"])
    .to_csv("books_cleaned.csv", index = False)
)
```
